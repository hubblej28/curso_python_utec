```python
import pandas as pd
import numpy as np
from sklearn.utils import shuffle
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/germaingarcia/Files_Lectures/refs/heads/main/PythonProgram/spotify_churn_dataset_es2.csv')
```

```python
df.head()
```

Mezclamos la poblacion

```python
poblacion = shuffle(df)
print(f"Tamaño de la población es: {len(poblacion)} registros\n")
```

Seleccionamos la muestra aleatoria

```python
muestra = poblacion.sample(n=800)
print(f"Tamaño de la pomuestra es: {len(muestra)} registros\n")
```

Unidad estadistica

```python
unidad = muestra.sample(n=1, random_state=5)
print("Unidad estadística (una fila de la muestra):")
display(unidad.T)
```
